```python
import pandas as pd
```

# Read CSV file and perform basic cleaning

```python
df= pd.read_csv("exoplanet_dataset.csv")
# find out the type of each column and drop all objects and rowid and kepid
df.dtypes
df_update= df.drop(['rowid', 'kepid', 'kepoi_name', 
                    'kepler_name', 'koi_pdisposition', 'koi_score',
                    'koi_tce_delivname'], axis=1)
df_update= df_update.dropna(axis='columns', how='all')
df_update= df_update.dropna()
df_update.head()
```

```python
y= df_update["koi_disposition"]
X= df_update.drop(columns= "koi_disposition")
# check if our selected feature has samples of atleast 2 classes
df_update.groupby('koi_disposition').count() 
```

# Create Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, random_state=1, shuffle=True)
```

# Pre-processing

```python
# scale data using MinMaxScaler and perform feature selection
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```
